```python
def studyPlanner():
    subjs=[]
    topics=[]
    for i in range(0,4):
        subjs.append(input("Enter a subject: "))
        topics.append(input("Enter a topic in that subject: "))
    return subjs,topics

def displayAll(s,t):
    print("All subjects and topics:")
    for i in range(0,4):
        print(f"{s[i]}: {t[i]}")

def search(s,t):
    find=input("What do you want to look for?: ")
    for i in range(0,4):
        if find == s[i]:
            return True
        elif find == t[i]:
            return True

def remove(s,t):
    rem=input("What is the name of the subject/topic you want to remove: ")
    for i in range(0,4):
        if rem==s[i]:
            s[i]="None"
        elif rem==t[i]:
            t[i]="None"
    
def menu():
    s,t=studyPlanner()
    while True:
        print("\n----- Study Planner -----")
        print('''1. Enter subjects \n2. Display subjects and topics
3. Search for subject/topic\n4. Remove subject/topic\n5. Exit''')
        ch=int(input("Enter your choice: "))
        if ch==1:
            s,t=subjectPlaner()
        elif ch==2:
            displayAll(s,t)
        elif ch==3:
            isIn=search(s,t)
            if isIn==True:
                print("In study plan")
            else:
                print("Not in study plan")
        elif ch==4:
            remove(s,t)
        elif ch==5:
            break
menu()
```

```python
def newNotes(cat):
    chcat= input("1. Materials\n2. Processes\n3. Safty\nEnter Name of catagory: ")
    note= input("Enter the note: ")
    cat[chcat].append(note)
    return cat

def display(cat):
    chcat= input("1. Materials\n2. Processes\n3. Safty\nName of catagory: ")
    lis=cat[chcat]
    print(f"\n{chcat} Notes:\n")
    for note in lis:
        print(f"- {note}")
        
    
    
def menu():
    mat=[]
    pro=[]
    saf=[]
    cat={"Materials": mat, "Processes": pro, "Safty": saf}
    while True:
        print("\n----- Note Organiser -----")
        print('''1. Add new notes \n2. Display all notes in a catagory
3. Delete a catagory/topic\n4. Display all catagories\n5. Exit''')
        ch=int(input("Enter your choice: "))
        if ch==1:
            cat=newNotes(cat)
        elif ch==2:
            display(cat)
        elif ch==3:
            del_cat = input('''Enter the name of the category to delete (Materials, Processes, Safety): ''')
            if del_cat in cat:
                cat[del_cat] = []  
                print(f"{del_cat} category has been deleted.")
            else:
                print("Category not found.")
        elif ch==4:
            print("\nCategories available:")
            for category in cat:
                print(f"- {category}")
        elif ch==5:
            break
menu()
```

```python
def testSum():
    scores=[]
    total=0
    while True:
        inpu=int(input("Enter test score or enter -1 to stop: "))
        if inpu==-1:
            break
        else:
            scores.append(inpu)
    for i in range(len(scores)):
        total+=scores[i]
    print(f'''The total score is {total}
The number of students is {len(scores)}
The average is {total/len(scores)}''')
testSum()
```

```python
def Display(inv):
    print("reagent inventory:\n")
    for mat, qu in inv.items():
        print(f"{mat}: {qu}")

def add(inv):
    nm= input("Enter the name: ")
    try:
        qu=float(input(f"Enter the quantity of{nm} in grams or ml: "))
        if nm in inv:
            print(f"{nm} already exists. Updating quantity.")
            inv[nm] += qu
        else:
            inv[nm] = qu
    except ValueError:
        print("Invalid quantity. Please enter a number.")

def ser(inv):
    nm= input("Enter the name: ")
    if nm in inv:
        print(f"The quantity of {nm} is {inv[nm]}.")
    else:
        print(f"{nm} is not found in the inventory.")

def redu(inv):
    nm= input("Enter the name: ")
    if nm in inv:
        usqu = float(input(f"Enter the quantity to reduce for {nm} (in grams or mL): "))
        if usqu > inv[nm]:
                print(f"Not enough {nm} in inventory to reduce by {usqu}.")
        else:
            inv[nm] -= usqu
    else:
        print(f"{nm} is not found in the inventory.")

def inven():
    inv={}
    while True:
        print("\n----- Material Invintory -----")
        print('''\n1. Display all\n2. Add
3. Search\n4. Reduce\n5. Exit''')
        ch=int(input("Enter your choice: "))
        if ch==1:
            Display(inv)
        elif ch==2:
            add(inv)
        elif ch==3:
            ser(inv)
        elif ch==4:
            redu(inv)
        elif ch==5:
            break
        else:
            print("Invalid choice. Please choose a valid option.")
inven()
```

```python
def val(shape):
    shape_type = input("Enter shape type (circle, square, rectangle): ").lower()
    if shape_type == "circle":
        radius = float(input("Enter the radius of the circle: "))
        shape["circle"] = radius
    elif shape_type == "square":
        side = float(input("Enter the side length of the square: "))
        shape["square"] = side
    elif shape_type == "rectangle":
        width = float(input("Enter the width of the rectangle: "))
        height = float(input("Enter the height of the rectangle: "))
        shape["rectangle"] = (width, height)
    else:
        print("Invalid shape type. Please enter 'circle', 'square', or 'rectangle'.")

def display(shape):
    if "circle" in shape:
        radius = shape["circle"]
        area = 3.14159 * radius ** 2
        print(f"Circle: Radius = {radius}, Area = {area}")
    else:
        print("Circle: Not entered.")
    
    if "square" in shape:
        side = shape["square"]
        area = side ** 2
        print(f"Square: Side = {side}, Area = {area}")
    else:
        print("Square: Not entered.")
    
    if "rectangle" in shape:
        width, height = shape["rectangle"]
        area = width * height
        print(f"Rectangle: Width = {width}, Height = {height}, Area = {area}")
    else:
        print("Rectangle: Not entered.")

def search(shape):
    shape_name = input("Enter shape to search for (circle, square, rectangle): ").lower()
    
    if shape_name == "circle" and "circle" in shape:
        radius = shape["circle"]
        area = 3.14159 * radius ** 2
        print(f"Circle: Radius = {radius}, Area = {area}")
    elif shape_name == "square" and "square" in shape:
        side = shape["square"]
        area = side ** 2
        print(f"Square: Side = {side}, Area = {area}")
    elif shape_name == "rectangle" and "rectangle" in shape:
        width, height = shape["rectangle"]
        area = width * height
        print(f"Rectangle: Width = {width}, Height = {height}, Area = {area}")
    else:
        print(f"{shape_name.capitalize()}: Not entered or invalid shape.")

def Geomhelp():
    shape = {}
    while True:
        print("\n----- Geometry Helper -----")
        print('''\n1. Enter values\n2. Display all
3. Search\n4. Exit''')
        ch=int(input("Enter your choice: "))
        if ch == 1:
            val(shape)
        elif ch == 2:
            display(shape)
        elif ch == 3:
            search(shape)
        elif ch == 4:
            break
        else:
            print("Invalid choice. Please choose a valid option.")
Geomhelp()
```
